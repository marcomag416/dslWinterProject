```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
n = np.array([125, 150, 300, 500, 1000])
score = np.array([4.850, 4.814, 4.771, 4.750, 4.738])
```

```python
plt.figure(figsize=(6,4))
plt.plot(np.log10(n), score, '-o')
#plt.xscale('log')
plt.yticks(fontsize=12)
plt.xticks(ticks=np.log10(n), labels=n, fontsize=12)
plt.xlabel("N estimators (log scale)", fontsize=14)
plt.ylabel("Average Eucliden distance", fontsize=14)
plt.tight_layout()
plt.savefig("rf_n_estiamators_comparison.pdf", format="pdf")
```
